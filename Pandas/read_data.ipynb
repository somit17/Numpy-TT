Reading Data from multiple / different sources

```python
import pandas as pd
from io import StringIO
import os
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
df=pd.read_json(StringIO(Data))
df
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient='records')
```

```python
##Read from URL
df_http = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
df_http.head(5)

try :
    df_http.to_csv("data/wine.csv")
    print(f"File Saved successfully - Done")
except Exception as e:
    print(f"Error saving file - {e}")
```
