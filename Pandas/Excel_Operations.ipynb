Understanding of loc and Iloc

Simple rule: Use loc when you know the name, use iloc when you know the position.

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['NYC', 'LA', 'Chicago']
}, index=['a', 'b', 'c'])

# Select by label
df.loc['a']  # Row with index 'a'
df.loc['a':'b']  # Rows 'a' to 'b' (inclusive)
df.loc['a', 'Name']  # Specific cell
df.loc[:, 'Age']  # All rows, 'Age' column

# Select by position
df.iloc[0]  # First row
df.iloc[0:2]  # First two rows (0 and 1, excludes 2)
df.iloc[0, 1]  # First row, second column
df.iloc[:, 1]  # All rows, second column
```

Accessing specified element

```python
df.at['a','Name']
df.iat[2,2]
```

| Method | Uses | Example |
|--------|------|---------|
| `at` | Labels (single value) | `df.at['c', 'Name']` |
| `iat` | Integer positions (single value) | `df.iat[2, 0]` |
| `loc` | Labels (can select multiple) | `df.loc['c', 'Name']` |
| `iloc` | Integer positions (can select multiple) | `df.iloc[2, 0]` |
