Pandas-DataFrame And Series

 It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

--Start-->

```python
import pandas as pd
```

Series - 1D Like Object  similar to column of table

```python
data = [1,2,3,4,5]
series = pd.Series(data)
print(f"Series - {series}") ##Return index and value and datatype
print(type(series))
```

Create series from dic

```python
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(f"Series - {series_dict}")
```

```python
data = [10,20,30]
indexes = ['a','b','c']

pd.Series(data=data,index=indexes)
```

DATAFRAME -

Create Dataframe from dic

```python
data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New York','Florida']
} #Must have same number of elements

df = pd.DataFrame(data)
print(f"Data - \n{df}") #basically like excel
print(f"{type(df)}")

```

Create Data frame from list of dic

```python
data=[
    {'Name':'Krish','Age':32,'City':'Bangalore'},
    {'Name':'John','Age':34,'City':'Bangalore'},
    {'Name':'Happy','Age':32,'City':'Bangalore'},
    {'Name':'Jack','Age':32,'City':'Bangalore'}
]
df = pd.DataFrame(data)
print(f"{df}")
```
