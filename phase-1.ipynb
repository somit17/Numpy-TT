## Numpy array and basics  - 

```python
import numpy as np
```

### Creating array from list


```python
arr_1d = np.array([1,2,3,4,5])
print("1D array : - ",arr_1d)
```

```python
arr_2d = np.array([[1,2,3],[4,5,7]])
print(f"2D array - {arr_2d}")
```

List vs Numpy Array

```python
py_list = [1,2,3]
print(f"Python list multiplication {py_list * 2}") #List is duplicated
```

```python
np_array = np.array([1,2,4])
print(f"Numpy multiplication {np_array * 2}")  #Values are multiplied -- Element wise multiplication

import  time
start  = time.time()
py_list = [i*2 for i in range(1000000)]
print(f"\n List operation time : {time.time()-start}")


start = time.time()
np_array = np.arange(1000000) * 2
print(f"\n Numpy operation time : {time.time()-start}")
```

Creating array from scartch

```python
zeros = np.zeros((3,4))
print(f"zeros array 3 Rows,4 Cols\n {zeros}")
```

```python
ones = np.ones((2,3))
print(f"One's array 2 Rows,4Cols\n {ones}")
```

```python
full = np.full((2,2),7)
print(f"Full array : - \n {full}")
```

```python
random = np.random.random((2,3))
print(f"Random array - \n{random}")
```

```python
sequence = np.arange(0,10,2)
print(f"Seq array - \n{sequence}")
```

Vector , Matrix and Tensor


```python
vector  = np.array([1,2,3])
matrix  = np.array([[1,2,3],
                    [4,5,6]])
print(f"Matrix- \n{matrix}")

tensor = np.array([[[1,2],[3,4]],
                   [[5,6],[7,8]]])
print(f"Tensor - \n{tensor}")  #Multi-dimensional array
```

Array Properties


```python
arr = np.array([[1,2,3],[4,5,6]])
print(f"Shape - {arr.shape}") #Gives rows and cols
print(f"Dimension  - {arr.ndim}")
print(f"Size  - {arr.size}") #Elements
print(f"DType  - {arr.dtype}") #Datatype
```
